 Load the dataset using pandas

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt  
#Load iris 
df=pd.read_csv("iris.csv")
```

```python
#to check rows and columns in a dataset we use shape function
print(f"Shape of Dataset: {df.shape}")
#to show first five rows of a dataset we use head
print(f"First Five Rows{df.head}")
```

```python
#to show column names we use 
print(f"Columns of Dataset are: {df.columns.tolist()}")
```

```python

# to show summary statistics we will use info and describe 
print(f"Dataset info: {df.info}")
print(f"Detail of dataset:  {df.describe}")
```

```python
#Histograms to show value distributions 

sns.histplot(df['SepalLengthCm'], kde=True)
plt.title('Histogram of Sepal Length')
plt.show()
sns.histplot(df['PetalLengthCm'], kde=True)
plt.title('Histogram of Petal Length')
plt.show()
sns.histplot(df['SepalWidthCm'], kde=True)
plt.title('Histogram of Sepal Width')
plt.show()
sns.histplot(df['PetalWidthCm'], kde=True)
plt.title('Histogram of Petal Width')
plt.show()
```

```python
#Here we are pointing outliers in a dataset by using boxplot.
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
df=pd.read_csv("iris.csv")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
plt.title("Box Plot to Identify Outliers")
#Here  in an orange boxplot of Sepal Width dots after whiskers showing outliers in a Sepal Width column. 
plt.show()

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

sns.scatterplot(data=df, x='PetalLengthCm', y='PetalWidthCm', 
                hue='Species', alpha=0.7, ax=ax1)
ax1.set_title('Petal: Better Separation')

sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', 
                hue='Species', alpha=0.7, ax=ax2)
ax2.set_title('Sepal: More Overlap')

plt.tight_layout()
plt.show()
```
